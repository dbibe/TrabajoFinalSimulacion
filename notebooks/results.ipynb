```python
import pandas as pd

results = pd.read_csv('../data/simulation_results.csv')

results.head()
```

```python
TEAL = '#34909d'
INFO = '#55c0ed'
DANGER = '#ba0034'
WARNING = '#f8b85e'
TEXT = '#35363b'
GRAY = '#d9dde3'
GRAY_DARKER = '#ced0d6'
GRAY_DARKEST = '#6f7082'

font = {
    'family': 'Lato',
    'color': TEXT,
    'weight': 'black',
}
```

## PPS

```python
import matplotlib.pyplot as plt

fig = plt.figure(facecolor='none', figsize=(10, 8))
ax = fig.add_subplot(111)
plt.title('PPS', fontdict=font)
plt.xlabel('(Junior, Sr) Engineers', fontdict=font)
plt.plot(results['label'], results['pps'], color=TEAL, label='PPS')
ax.set_facecolor('none')
plt.show()
```

### Cost


```python
results['loss'] = results['jr_cost'] * (results['max_pto_jr'] / 100) + results['sr_cost'] * (
        results['max_pto_sr'] / 100)

fig = plt.figure(facecolor='none', figsize=(10, 8))
ax = fig.add_subplot(111)
plt.title('Cost vs Loss due to Idle', fontdict=font)
plt.xlabel('(Junior, Sr) Engineers', fontdict=font)
plt.plot(results['label'], results['jr_cost'] + results['sr_cost'], color=TEAL, label='Cost')
plt.plot(results['label'], results['loss'], color=DANGER, label='Loss', linestyle='--')
plt.legend()
ax.set_facecolor('none')
plt.show()
```
